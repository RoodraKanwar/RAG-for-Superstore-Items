```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("food_items_list_with_macros.csv")
df.head(5)
```

```python
df = df.replace(np.nan, 'unknown')
df['product_size'] = df['product_size'].str.replace("ea", "eggs")
df['product_size'] = df['product_size'].str.replace("$6.61/1kg", "2.35 kg")
df['product_size'] = df['product_size'].str.replace("12x100.0 g", "Pack of 12 yogurts, each 100g")
df['unit_measure'] = df['unit_measure'].str.replace("1ea", "egg")
df['protein'] = df['protein'].str.replace("7g39g", "7g/39g")
df[['protein', 'macro_weight']] = df['protein'].str.split('/', expand=True)
df['carbs'] = df['carbs'].str.split('/').str[0]
df['fats'] = df['fats'].str.split('/').str[0]

for col in ['protein', 'carbs', 'fats']:
    df[col] = df[col].astype(str).str.replace(r'(?<!g)$', 'g', regex=True)
df['macro_weight'] = df['macro_weight'].str.replace("2ea", "2 eggs")
```

```python
df["text"] = (
    "Brand is " + df['brand'] + ". " +
    "Product is " + df['product_name'] + ". " +
    "Base Price is $" + df['base_price'].astype(str) + ". " +
    "Size is " + df['product_size'] + ". " +
    "Unit Price is $" + df['unit_price_value'].astype(str) + " per " + df['unit_measure'] + ". " +
    "It contains " + df['protein'] + " of protein, " + df['carbs'] + " of carbs, and " + df['fats'] + " of fats per " + df['macro_weight'] + "."
)
```
